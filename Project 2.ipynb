```python
# Dependencies
import pandas as pd

from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Import the used car dataset
car_reviews_df = pd.read_csv('Resources/reviews_data.csv')
car_reviews_df.head()
```

```python
#Set Review_ID as index
car_reviews_df = car_reviews_df.set_index('Review_ID')

#Drop all null values
car_reviews_df = car_reviews_df.dropna()
```

```python
vehicle_titles = car_reviews_df['Vehicle_Title']

#Creating lists for year, make, and model to be the considered the vehicle name.
Vehicle_Name = []

for vehicle in vehicle_titles:
    
    #Extract make, year, and model and append to lists.
    title_list = vehicle.split()
    name = pd.Series([title_list[0], title_list[1], title_list[2]])
    vehicle_name = name.str.cat(sep = ' ')
    
    Vehicle_Name.append(vehicle_name)

car_reviews_df['Vehicle_Name'] = Vehicle_Name

car_reviews_df
```

```python
vehicle_ratings = car_reviews_df[['Rating','Vehicle_Name']]
vehicle_ratings_summary = vehicle_ratings.groupby('Vehicle_Name').mean()

vehicle_ratings_summary
```

```python
# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
#Creating lists of all names, prices, stock type, reviews, and review counts for new toyotas within 100 miles of area code 30332.
Names = []
Prices = []
Stock_Types = []
Dealers = []
CarShop_Distances = []

# Loop through returned results
for x in range(1, 9):
    
    s = pd.Series(['https://www.cars.com/shopping/results/?page=',str(x),'&page_size=100&dealer_id=&keyword=&list_price_max=&list_price_min=&makes[]=toyota&maximum_distance=200&mileage_max=&sort=best_match_desc&stock_type=new&year_max=&year_min=&zip=30332'])
    url = s.str.cat()
    browser.visit(url)
    
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    
    # Examine the results and look for a div with the class 'vehicle-details'
    results = soup.find_all('div', class_='vehicle-details')
    
    for result in results:
    
        # Retrieve the data
        name = result.find('h2', class_='title').text
        price = result.find('span', class_='primary-price').text
        dealer = result.find('div', class_='dealer-name').strong.text
        distance = result.find('div', class_='miles-from').text
        stock_type = result.find('p', class_='stock-type').text
    
    
        # Append to respective list
        Names.append(name)
        Prices.append(price)
        Dealers.append(dealer)
        CarShop_Distances.append(distance)
        Stock_Types.append(stock_type)
```

```python
new_car_dict = {
    'Name': Names,
    'Price($)': Prices,
    'StockType': Stock_Types,
    'Distance from 30332': CarShop_Distances,
    'Dealership': Dealers
}

new_cars_df = pd.DataFrame(data = new_car_dict)
new_cars_df.head()
```

```python
#Transforming and cleaning the data 

#Import dataset
new_cars_df = pd.read_csv('Excel Files/new_cars_data.csv')

used_cars_df = pd.read_csv('Excel Files/used_cars_data.csv')
```

```python
#Remove any not priced new vehicles
new_cars_df = new_cars_df.loc[new_cars_df['Price($)'] != 'Not Priced',:]

#Remove the $ in price and set as numeric values
Prices = new_cars_df['Price($)']
Prices_Num = []

for price in Prices: 
    price_new = price[1:]
    price_list = price_new.split(',')
    price_cat = pd.Series([price_list[0], price_list[1]])
    price_num = price_cat.str.cat()

    Prices_Num.append(price_num)

new_cars_df['Price($)'] = Prices_Num
new_cars_df['Price($)'] = pd.to_numeric(new_cars_df['Price($)'])
```

```python
#Remove any not priced used vehicles
used_cars_df = used_cars_df.loc[used_cars_df['Price($)'] != 'Not Priced',:]

#Remove the $ in price and set as numeric values
Prices = used_cars_df['Price($)']
Prices_Num = []

for price in Prices: 
    price_new = price[1:]
    price_list = price_new.split(',')
    price_cat = pd.Series([price_list[0], price_list[1]])
    price_num = price_cat.str.cat()

    Prices_Num.append(price_num)

used_cars_df['Price($)'] = Prices_Num
used_cars_df['Price($)'] = pd.to_numeric(used_cars_df['Price($)'])
```

```python
#Merge the new and used cars datasets
cars_df = used_cars_df.append(new_cars_df,ignore_index=True)
cars_df.head()

#Rename the Price and Distance Column 
cars_df = cars_df.rename(columns = {'Price($)':'Price', 'Distance from 30332': 'Distance_from_30332'})
cars_df.head()
```

```python
#Extract just the integer portion of the distance from area code 30332
Distances = cars_df['Distance_from_30332']
Distances_Num = []

for distance in Distances: 
    distance_list = distance.split(' ')
    distance_new = distance_list[0]
    
    try:
        distance_comma = distance_new.split(',')
        distance_combine = pd.Series([distance_comma[0],distance_comma[1]])
        distance_num = distance_combine.str.cat()

        Distances_Num.append(distance_num)
    
    except IndexError:
        Distances_Num.append(distance_new)
        
    
cars_df['Distance_from_30332'] = Distances_Num
cars_df.head()
```

```python
#Remove all rows with an empty distance value 
cars_df = cars_df.loc[cars_df['Distance_from_30332'] != '',:]

#Set datatype for Distance to be int64
cars_df['Distance_from_30332'] = pd.to_numeric(cars_df['Distance_from_30332'])
```

```python
vehicle_titles = cars_df['Name']

#Creating lists for year, make, and model to be the considered the vehicle name.
Year = []
Make = []
Model = []

for vehicle in vehicle_titles:
    
    #Extract make, year, and model and append to lists.
    vehicle_list = vehicle.split()

    year = vehicle_list[0]
    make = vehicle_list[1]
    model = vehicle_list[2]
    
    Year.append(year)
    Make.append(make)
    Model.append(model)

cars_df['Year'] = Year
cars_df['Make'] = Make
cars_df['Model'] = Model
```

```python
#Remove duplicates and null rows 
cars_df = cars_df.drop_duplicates()
cars_df =cars_df.dropna()

cars_df.head()
```
